```python
str ="car, car, bike, car, bike, car, bike"
```

```python
arr = str.split(", ")
print(arr)
```

```python
word_count = {}
for word in arr:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```
